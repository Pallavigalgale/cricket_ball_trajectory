```python
import os
import cv2
import pandas as pd
import numpy as np
from ultralytics import YOLO
from scipy.interpolate import interp1d

# 1. Setup
video_input_dir = '/content/drive/MyDrive/edgefleet_assessment/test_data/25_nov_2025'
output_dir = '/content/final_submission_results'
os.makedirs(output_dir, exist_ok=True)

# 2. Model Load
model = YOLO('/content/drive/MyDrive/edgefleet_assessment/models/best.pt')

def smooth_path(points):
    if len(points) < 3: return points
    # Points madhla noise kami karnyasaathi simple moving average
    pts = np.array(points)
    window = 3
    smoothed = np.copy(pts)
    for i in range(window, len(pts)-window):
        smoothed[i] = np.mean(pts[i-window:i+window], axis=0)
    return [tuple(x) for x in smoothed.astype(int)]

for video_name in os.listdir(video_input_dir):
    if video_name.endswith(('.mp4', '.avi', '.mov')):
        video_path = os.path.join(video_input_dir, video_name)
        video_base_name = os.path.splitext(video_name)[0]

        cap = cv2.VideoCapture(video_path)
        w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
        video_writer = cv2.VideoWriter(os.path.join(output_dir, f"{video_base_name}_ultra_pro.mp4"),
                                     cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))

        path_points = []
        frame_list = []

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret: break

            frame_idx = int(cap.get(cv2.CAP_PROP_POS_FRAMES)) - 1

            # ROI adjustment
            roi_y1, roi_y2 = int(h * 0.1), int(h * 0.95)
            roi_frame = frame[roi_y1:roi_y2, :]

            results = model.predict(roi_frame, conf=0.35, verbose=False)

            ball_found_this_frame = False
            for r in results:
                boxes = r.boxes.xywh.cpu().numpy()
                if len(boxes) > 0:
                    x, y, wb, hb = boxes[0]
                    actual_y = y + roi_y1
                    center = (int(x), int(actual_y))
                    path_points.append(center)

                    # Green circle with glow effect
                    cv2.circle(frame, center, 6, (0, 255, 0), -1)
                    cv2.circle(frame, center, 10, (0, 255, 0), 2)

                    frame_list.append([frame_idx, round(float(x), 1), round(float(actual_y), 1), 1])
                    ball_found_this_frame = True
                    break

            if not ball_found_this_frame:
                frame_list.append([frame_idx, -1, -1, 0])

            # Smart Path Drawing
            if len(path_points) > 1:
                smoothed_pts = smooth_path(path_points)
                for i in range(1, len(smoothed_pts)):
                    # Trajectory line with Neon Red look
                    cv2.line(frame, smoothed_pts[i-1], smoothed_pts[i], (0, 0, 255), 3, cv2.LINE_AA)

            video_writer.write(frame)

        cap.release()
        video_writer.release()

        # Save CSV
        pd.DataFrame(frame_list, columns=['frame', 'x', 'y', 'visible']).to_csv(
            os.path.join(output_dir, f"{video_base_name}_annotations.csv"), index=False)
        print(f"Final Ultra Video: {video_name}")

print("\n--- Project complete at Maximum Quality! ---")
```

```python
import os
import shutil

# 1. Define paths
submission_base = '/content/submission'
final_results_source = '/content/final_submission_results'
model_source = '/content/drive/MyDrive/edgefleet_assessment/models/best.pt'

# 2. Clean start: Delete the old submission folder if it exists to avoid confusion
if os.path.exists(submission_base):
    shutil.rmtree(submission_base)

# 3. Create fresh GitHub repository structure
folders = ['code', 'annotations', 'results', 'model']
for folder in folders:
    os.makedirs(os.path.join(submission_base, folder), exist_ok=True)

# 4. Copy Annotations (CSV files)
csv_count = 0
for file in os.listdir(final_results_source):
    if file.endswith('.csv'):
        shutil.copy(os.path.join(final_results_source, file), os.path.join(submission_base, 'annotations'))
        csv_count += 1

# 5. Copy Processed Videos (Ultra Pro versions for best quality)
video_count = 0
for file in os.listdir(final_results_source):
    if 'ultra_pro' in file:
        shutil.copy(os.path.join(final_results_source, file), os.path.join(submission_base, 'results'))
        video_count += 1

# 6. Copy Model weight file
if os.path.exists(model_source):
    shutil.copy(model_source, os.path.join(submission_base, 'model'))
    print("Model file copied successfully.")
else:
    print("Warning: Model file not found at source path!")

print(f"Structure Ready: {csv_count} annotations and {video_count} videos organized.")
```
